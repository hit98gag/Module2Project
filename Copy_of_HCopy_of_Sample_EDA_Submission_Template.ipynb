## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

import pandas as pd
import numpy as np

# Assume df is already loaded

# 1. Convert date columns to datetime
date_cols = [
    'PQ First Sent to Client Date',
    'PO Sent to Vendor Date',
    'Scheduled Delivery Date',
    'Delivered to Client Date',
    'Delivery Recorded Date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# 2. Handle missing values

# Shipment Mode - fill missing with 'Unknown'
df['Shipment Mode'].fillna('Unknown', inplace=True)

# Dosage - fill missing with 'Not Specified'
df['Dosage'].fillna('Not Specified', inplace=True)

# Line Item Insurance (USD) - fill missing with 0
df['Line Item Insurance (USD)'].fillna(0, inplace=True)

# 3. Convert numeric columns stored as object to numeric (if needed)
# Columns: Weight (Kilograms), Freight Cost (USD) are object type but should be numeric

df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'], errors='coerce')
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'], errors='coerce')

# 4. Remove duplicates if any
df.drop_duplicates(inplace=True)

# 5. Strip whitespace from string columns to avoid hidden issues
str_cols = df.select_dtypes(include='object').columns
for col in str_cols:
    df[col] = df[col].str.strip()

# 6. Final sanity checks
print("Missing values after cleaning:")
print(df.isnull().sum())

print("\nData types after conversion:")
print(df.dtypes)

print("\nDataset shape:", df.shape)

# Now df is cleaned and ready for further analysis

````
